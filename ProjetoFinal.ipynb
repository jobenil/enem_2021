Históricamente o Brasil vêm realizando através do INEP -  Instituto Nacional de Estudos e Pesquisas Educacionais Anísio Teixeira uma série de provas de âmbito nacional como o **SAEB** e **Prova Brasil** com vistas a avaliação do Sistema Nacional de Ensino Básico e Fundamental. Tais avaliações tem por objetivo mensurar a qualidade do ensino efetuado pelos diferentes entes federativos, quer no âmbito público quanto no âmbito privado.
Pretende-se que tais avaliações prossibilitem aos **administradores** das diferentes redes de ensino reflitam e redirecionem **ações** no sentido de efetuar melhorias em seus respectivos níveis de responsabilidade.
O **ENEM** - Exame Nacional do Ensino Médio, é uma prova realizada pelo Ministério da Educação desde 1998 com objetivo de avaliar a qualidade do ensino médio praticado no Brasil e que a partir de 2009 amplia o escopo visando a seleção de candidatos às universidades federais do país.
Desde sua criação emprega na elaboração das notas dos candidatos uma metodologia denominada TRI - Teoria da Resposta ao Item. A referida teoria pretende aferir a partir da análise sistemática de cada questão da prova, um nível denominado nível de proficiência do candidato.
O presente trabalho pretende levantar como e o quanto as diferentes áreas do conhecimento afetam o resultado final do aluno e verificar quais fatores socioeconômicos influênciam esses resultados. Para esse fim selecionaremos os dados referentes aos candidatos no estado de Goiás para o exame realizado em 2021.

#  <a name="secao_1"> 1. Teoria da Resposta ao Item </a>
[voltar ao sumário](#sumario)

A Teoria da Resposta ao Item foi inicialmente proposta entre 1950 e 1960 por três pesquisadores independentes [Frederic M. Lord](https://en.wikipedia.org/wiki/Frederic_M._Lord),  [Georg Rach](https://en.wikipedia.org/wiki/Georg_Rasch) e [Paul Lazarsfeld](https://en.wikipedia.org/wiki/Paul_Lazarsfeld) e tinha por objetivos superar algumas dificuldades de interpretação dos resultados dos testes de proficiência.
De uma maneira geral, podemos dizer que os testes de proficiência acadêmica, sobretudo os testes de **"lápis e papel"** possuem mais características **subjetivas** do que **objetivas** pois procuram medir algum tipo de proficiência cognitiva chamadas de **traços latentes**. No presente caso, o ENEM procura medir as proficiências em Matemática e suas Tecnologias; Linguagens, Códigos e suas Tecnologias; Ciências da Natureza e suas Tecnologias; e Ciências Humanas e suas Tecnologias. 
A métodologia tradicionalmente utilizada em avaliações de ensino e educação é chamada de TCL - Teoria Clássica dos Testes a qual considera o número total de acertos em um teste como base para concepção de suas validações, desconsiderando **peculiaridades** inerentes a cada uma das suas questões, sendo que essencialmente os **traços latentes** apresentam-se mais como numa escala **ordinal** do que numa escala de **razão** o que "teóricamente" invalidaria as conclusões para as quais os citados testes serviriam.
Na TRI ficam estabelecidos a identificação das características - **discriminação, dificuldade e aleatoriedade** para cada **item** da prova, o que permite calcular suas probabilidades de acertos dado um certo nível de proficiência do respondente sendo essa última uma característica **"imanente"** do mesmo e portanto indicador **"inequivoco"** da habilidade. É com base nessas características que diversas instituições como o College Board que aplica o SAT Schollastic Aptitude Test utilizado pelos jovens americanos acessarem o ensino superior, a OCDE para execução do PISA e o INEP para execução de suas avaliações educacionais.

No caso do ENEM cada uma das questões é préviamente calibrada através da aplicação das mesmas a uma população de alunos do ensino médio selecionados aleatóriamente nas diversas escolas do país. Posteriormente é calculada função característica do item conforme a função descrita abaixo:$$p_i(\theta)= c_i + \frac{1-c_i}{1 + e^{-a_i(\theta - b_i}}$$

onde,
- $a_i$ é o índice de discriminação do item i
- $b_i$ é o índice de dificuldade do item i
- $c_i$ é um fator de aleatóriedade do item i, considerando a possibilidade de um aluno acertar no chute.

$\theta$ indica o nível de proficiência do aluno, que no caso do ENEM tem média 500 e desvio padrão 100 considerando uma escala na distribuição normal.


As questões assim calibradas são colocadas em um banco chamado de Banco de Questões as quais serão selecionadas para compor a prova do ENEM.

A figura abaixo apresenta a curva característica de um item segundo a TRI.

![](cci_exemplo.png)

Observamos todos os citados parâmetros a, b, c e $\theta$ representado aqui como os níveis de proficiência em uma escala que vai de $-\infty$ a $\infty$. No caso da proficiência estabelecida no ENEM uma escala normalizada com média 500 e desvio padrão 100.

Muito embora saibamos que as escalas de proficiências obtidas a partir da TRI para cada uma das áreas de conhecimento não sejam **"teóricamente"** comparáveis, nesta seção proporemos além da análise descritiva, uma analise comparativa empregando para isso a técnica estatística denominada **ANOVA** - Análise de Variância ou a aplicação de algum teste estatística para médias quando a comparação se der entre apenas dois grupos.
Podemos pensar na utilização destas metodologias por consideramos que os resultados de proficiência foram normalizados por construção segundo os mesmos valores de média e desvio padrão. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import researchpy as rp
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
arq = "./DADOS/MICRODADOS_ENEM_2021_FILTRADO.csv"
base_trat = pd.read_csv(arq)
```

```python
base_trat.head().T
```

### Comparação das médias nas notas de Matemática; Linguagens/Códigos; Ciências da Natureza e Ciências Humanas

```python
notas = base_trat[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']]

notas_temp = notas.melt(id_vars=[], var_name = "Tratamentos", value_name = "Nota")
ax = sns.boxplot(data=notas, orient='h', palette="Set2")
ax = sns.displot(data=notas_temp, x='Nota', hue = 'Tratamentos', palette='Set2', kind='kde')
```

Com base nos gráficos acima notamos que possívelmente há uma pequena diferença entre as áreas de Ciências da Natureza, Ciências Humanas, Linguagem e Matemática e uma diferença mais significativa da nota de Redação para as demais. A fim da confirmação dessa hipótese empregaremos uma Análise de Variância com vistas a identificar se há evidências estatísticas de diferença entre elas e a aplicação de um teste de TUKEY com o objetivo de identificar onde as possíveis diferenças se encontram.

```python
rp.summary_cont(notas)
```

```python
fvalue, pvalue = stats.f_oneway(base_trat['NU_NOTA_CN'], base_trat['NU_NOTA_CH'], base_trat['NU_NOTA_LC'], base_trat['NU_NOTA_MT'], base_trat['NU_NOTA_REDACAO'])
print(round(fvalue,4), round(pvalue,4))
```

```python
model = ols('Nota ~ Tratamentos', data=notas_temp).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

Com um p-value de 0.0 pode-se dizer que há diferenças estatísticas entre as médias das notas obtidas. O teste de Tukey abaixo irá identificar onde essas diferenças se encontram.

```python
from bioinfokit.analys import stat
res = stat()
res.tukey_hsd(df=notas_temp, res_var='Nota', xfac_var='Tratamentos', anova_model='Nota ~ C(Tratamentos)')
res.tukey_summary
```

Através da tabela apreende-se que há diferenças estatísticas **muito significativas** pois em todas as comparações os **p-valores** foram inferiores a **1%**. Destacamos as diferenças obtidas entre as notas de Redação e Ciências da Natureza (134 pontos), Redação e Linguagem (120 pontos) e Redação e Ciências Humanas (105 pontos).
Note que, embora vislumbrassemos essas diferenças através da inspeção visual dos gráficos, apenas pudemos ter certeza dessas diferenças após a execução de um teste estatístico de hipóteses como o que realizamos.
Um ponto importânte a se destacar neste caso é que para a obtenção das notas de Redação não seguiu-se a mesma metodologia que a utilizada para obtenção das demais.

```python
# Altera os campos TP_FAIXA_ETARIA, TP_SEXO, TP_ESTADO_CIVIL, TP_COR_RACA, TP_DEPENDENCIA_ADM_ESC,
# e PONTUACAO deixando-as como uma string mais explicativa.
base_trat['TP_FAIXA_ETARIA'].replace({1:'Menor de 17 anos', 2:'17 anos', 3:'18 anos', 4:'19 anos',
                                      5:'20 anos', 6:'21 anos', 7:'22 anos', 8:'23 anos', 9:'24 anos',
                                      10:'25 anos', 11:'Entre 26 e 30 anos', 12:'Entre 31 e 35 anos',
                                      13:'Entre 36 e 40 anos', 14:'Entre 41 e 45 anos', 15:'Entre 46 e 50 anos',
                                      16:'Entre 51 e 55 anos',17:'Entre 56 e 60 anos', 18:'Entre 61 e 65 anos',
                                      19:'Entre 66 e 70 anos', 20:'Maior de 70 anos'},inplace= True)

base_trat['TP_SEXO'].replace({"F": 'Feminino', "M": 'Masculino'}, inplace= True)

base_trat['TP_ESTADO_CIVIL'].replace({0:'Não informado', 1:'Solteiro(a)',
                                      2:'Casado(a)/Mora com companheiro(a)',
                                      3:'Divorciado(a)/Desquitado(a)/Separado(a)',
                                      4:'Viúvo(a)'}, inplace= True)

base_trat['TP_COR_RACA'].replace({0:'Não declarado', 1:'Branca', 2:'Preta', 3:'Parda',
                                  4:'Amarela', 5:'Indígena', 6:'Não dispõe da informação'}, inplace= True)

base_trat['TP_DEPENDENCIA_ADM_ESC'].replace({1:'Federal', 2:'Estadual', 3:'Municipal', 4:'Privada'}, inplace= True)

base_trat['PONTUACAO'].replace({0:'DE', 1:'C2', 2:'C1', 3:'B1', 4:'B2', 5:'A'}, inplace= True)
```

```python
def sumario_agrupado(nota, var2):
    # Estatística Descritiva das notas separados por var2
    print()
    print('Sumário de ', nota, ' ', 'por ',var2, '.')
    print(rp.summary_cont(base_trat[nota].groupby(base_trat[var2])))
    print()
```

```python
variaveis = ['TP_FAIXA_ETARIA', 'TP_SEXO', 'TP_ESTADO_CIVIL', 'TP_COR_RACA', 'TP_DEPENDENCIA_ADM_ESC', 'PONTUACAO']
notas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']

for nota in notas:
    for var in variaveis:
        sumario_agrupado(nota, var)      
```

Para todas as comparações efetuadas anteriormente os valores das médias aparentam ser estatísticamente diferentes. Com vistas a confirmação destas hipóteses procederemos a execução de uma análise de variância comparando as respectivas notas e variáveis comparativas.

```python
compara = ['NU_NOTA_CN ~ TP_FAIXA_ETARIA', 'NU_NOTA_CN ~ TP_SEXO', 'NU_NOTA_CN ~ TP_ESTADO_CIVIL', 'NU_NOTA_CN ~ TP_COR_RACA', 'NU_NOTA_CN ~ TP_DEPENDENCIA_ADM_ESC', 'NU_NOTA_CN ~ PONTUACAO',
        'NU_NOTA_CH ~ TP_FAIXA_ETARIA', 'NU_NOTA_CH ~ TP_SEXO', 'NU_NOTA_CH ~ TP_ESTADO_CIVIL', 'NU_NOTA_CH ~ TP_COR_RACA', 'NU_NOTA_CH ~ TP_DEPENDENCIA_ADM_ESC', 'NU_NOTA_CH ~ PONTUACAO',
        'NU_NOTA_LC ~ TP_FAIXA_ETARIA', 'NU_NOTA_LC ~ TP_SEXO', 'NU_NOTA_LC ~ TP_ESTADO_CIVIL', 'NU_NOTA_LC ~ TP_COR_RACA', 'NU_NOTA_LC ~ TP_DEPENDENCIA_ADM_ESC', 'NU_NOTA_LC ~ PONTUACAO',
        'NU_NOTA_MT ~ TP_FAIXA_ETARIA', 'NU_NOTA_MT ~ TP_SEXO', 'NU_NOTA_MT ~ TP_ESTADO_CIVIL', 'NU_NOTA_MT ~ TP_COR_RACA', 'NU_NOTA_MT ~ TP_DEPENDENCIA_ADM_ESC', 'NU_NOTA_MT ~ PONTUACAO',
        'NU_NOTA_REDACAO ~ TP_FAIXA_ETARIA', 'NU_NOTA_REDACAO ~ TP_SEXO', 'NU_NOTA_REDACAO ~ TP_ESTADO_CIVIL', 'NU_NOTA_REDACAO ~ TP_COR_RACA', 'NU_NOTA_REDACAO ~ TP_DEPENDENCIA_ADM_ESC', 'NU_NOTA_REDACAO ~ PONTUACAO']

def anova(compara):
    model = ols(compara, data=base_trat).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    print(anova_table)

for comp in compara:
    print(comp)
    anova(comp)
    print()
    print()
```
